# Student Performance Analysis
This notebook analyzes student performance data using pandas and matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Step 1: Import Libraries
In this step, we import the required libraries for data analysis and visualization.

```python
data = pd.read_csv('student-mat.csv',sep=';')
data
```

## Step 2: Load Dataset
The dataset is loaded using pandas. The semicolon (`;`) is used as the separator.

```python
data.head()
```

## Step 3: Display First Rows of Dataset
We use `head()` to view the first few rows of the dataset.

## Step 4: Check Missing Values
We check whether the dataset contains null values.

```python
data.isnull()
```

## Step 5: Count Missing Values
We calculate the total number of missing values in each column.

```python
data.isnull().sum()
```

## Step 6: Check Duplicate Records
We check for duplicate rows in the dataset.

```python
data.duplicated().sum()
```

## Step 7: Check Dataset Shape
We find the number of rows and columns in the dataset.

```python
data.shape
```

## Step 8: Check Data Types
We inspect the datatype of each column.

```python
data.dtypes
```

## Step 9: Dataset Information
We display detailed information about the dataset including non-null counts.

```python
data.info()
```

## Step 10: Calculate Average Final Grade
We calculate the average value of the final grade column (`G3`).

```python
mean = data['G3'].mean()
print(f"Mean: {mean}")
```

## Step 11: Count Students Scoring Above 15
We count how many students scored more than 15 marks in the final grade.

```python
higher = (data['G3']>15).sum()
higher
```

## Step 12: Correlation Between Study Time and Performance
We calculate the correlation between study time and final grades.

```python
data[['studytime','G3']].corr()
```

## Step 13: Scatter Plot Analysis
We create a scatter plot to visualize the relationship between study time and performance.

```python
plt.scatter(data['studytime'], data['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Performance")
plt.show()
```

## Step 14: Average Performance by Gender
We calculate the average final grade for each gender.

```python
data.groupby('sex')['G3'].mean()
```

## Step 15: Bar Chart Visualization
We visualize average student performance by gender using a bar chart.

```python
import matplotlib.pyplot as plt

data.groupby('sex')['G3'].mean().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.title("Average Performance by Gender")
plt.show()
```
